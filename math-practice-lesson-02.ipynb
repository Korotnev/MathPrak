### Гармонические колебания математического маятиника


Задаются уравнением $\alpha'' + w^2 \alpha = 0$, где $\alpha$ - угол отклонения. 


![](http://ru.solverbook.com/my_images/pic1103.png)



```python
import matplotlib.pyplot as plt
import numpy as np

# Константы, начальные параметры
gravit_acc = 9.8
length = 1
alpha0 = 0.01


# Уравнение для alpha(t):=
def alpha(t):
    return alpha0*np.cos(np.sqrt(gravit_acc/length)*t)



# Рисуем график
t = np.linspace(0, 300)

plt.plot(t, alpha(t), color = 'green')
plt.title('Грфик alpha(t)')

plt.grid(True)
plt.show()
```

### Task 2.2 

Требуется построить для этой системы графики

* положения (угла маятника $\alpha(t)$)
* скорости
* ускорения



```python
### Task 2.2. Solution:
import matplotlib.pyplot as plt
import numpy as np
e=2.71
m=5
k=0.1
n=8
alpha0 = int((n+3)%4 * 10 / 3) * 0.001
l=2*n

w=np.sqrt(gravit_acc/l)

r=k/2/m

def alpha(t):
    return (alpha0*np.cos(w*t)*e**(-1*r*w*t))

t = np.arange(0, 500, 0.02)

plt.plot(t, alpha(t), color = 'green')
plt.title('Грфик alpha(t)')

plt.grid(True)
plt.show()

def velocity_angle(t):
    i=-1*r*w*t
    i=e**(i)
    i=i*(-1*r*w*t*alpha0*np.cos(w*t)-w*alpha0*np.sin(w*t))
    return i

t = np.arange(0, 50, 0.02)
t = np.arange(0, 1000, 0.02)
plt.plot(t, velocity_angle(t), color = 'm')
plt.title('Грфик w(t)')

plt.grid(True)
plt.show()

def acc_angle(t):
    i=-1*r*w*t
    i=e**(i)
    i=i*(r*w*t*alpha0*np.cos(w*t)-w*alpha0*np.sin(w*t)+-1*r*w*t*alpha0*np.cos(w*t)-w*alpha0*np.sin(w*t))
    return i

t = np.arange(0, 50, 0.02)
t = np.arange(0, 1000, 0.02)
plt.plot(t, acc_angle(t), color = 'm')
plt.title('Грфик E(t)')

plt.grid(True)
plt.show()

























```
